```python
import requests
from bs4 import BeautifulSoup

# URL of the page
url = 'https://flatreviews.com/dubai/reviews/581'

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the review date elements
review_dates = soup.find_all('div', class_='fr-review-date')
person_class = soup.find_all('div', class_='fr-review-user-stats')
district_name = soup.find_all('div', class_=''
# Extract the review dates
for date in review_dates:
    print(f"Name {date.text.strip()}")
for person in person_class:
    print(person.text.strip())

# Find all the review content elements
positive_psides = soup.select('div.fr-review-posnegs:nth-of-type(4) div.large-auto')

# Extract the review contents
for psides in positive_psides:
    print('Positives:', psides.text.strip())

negative_nsides = soup.select('div.fr-review-posnegs:nth-of-type(5) div.large-auto')
for nsides in negative_nsides:
    print('Negatives:', nsides.text.strip())

review_texts = soup.select('div.fr-review-text.show-for-large')
for texts in review_texts:
    print('Review text:', texts.text.strip())

overall_rating = soup.select('.show-for-large div.fr-location-average-score')
for rating in overall_rating:
    print('Overall rating:', rating.text.strip())

parameters_rating = soup.select('.material-icons:not(.star-none)')

# Function to convert stars to numbers
def convert_stars_to_numbers(stars):
    star_mapping = {
        'star': '1',
        'star_half': '0.5',
        'star_border': '0'
    }
    return star_mapping.get(stars, 'Unknown')

# Extract and display parameters rating
for rating in parameters_rating:
    star_class = rating.get('class')
    if star_class:
        star_icon = star_class[-1]
        number = convert_stars_to_numbers(star_icon)
        print('Parameters rating:', number)
```
